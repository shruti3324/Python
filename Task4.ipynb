```python
import matplotlib.pyplot as plt
```

```python
year = list(range(11))  

Kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2000, 2100]
Sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0, 0]
Rohit = [0, 200, 500, 900, 1400, 1600, 1550, 1300, 800, 50, 0]

plt.plot(year, Kohli, label="Virat Kohli")
plt.plot(year, Sehwag, label="Virender Sehwag")
plt.plot(year, Rohit, label="Rohit Sharma")

plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
plt.grid(True)
plt.show()
```

```python
months = list(range(11))  

Electronics = [1000, 1500, 1800, 2000, 2500, 2300, 2200, 2100, 1900, 1700, 1600]
Clothing = [500, 700, 900, 1200, 1600, 2000, 1900, 1700, 1500, 1400, 1300]
Groceries = [800, 900, 1000, 1200, 1300, 1400, 1350, 1300, 1200, 1100, 1000]

plt.plot(months, Electronics, label="Electronics")
plt.plot(months, Clothing, label="Clothing")
plt.plot(months, Groceries, label="Groceries")

plt.xlabel("Month")
plt.ylabel("Sales (in units)")
plt.title("Monthly Sales Comparison")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.style.use('grayscale')

months = list(range(11)) 

Organic = [500, 800, 1000, 1300, 1600, 1800, 1700, 1600, 1500, 1400, 1300]
Paid = [200, 400, 700, 900, 1000, 1100, 1050, 950, 800, 600, 400]
Referral = [100, 200, 300, 500, 700, 750, 700, 650, 600, 550, 500]

plt.plot(months, Organic, label="Organic Traffic", linewidth=2)
plt.plot(months, Paid, label="Paid Traffic", linestyle='--', linewidth=2)
plt.plot(months, Referral, label="Referral Traffic", linestyle='-.', linewidth=2)

plt.xlabel("Month")
plt.ylabel("Visitors")
plt.title("Website Traffic Sources Over Time")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.style.use('dark_background')
months = list(range(11))

Mumbai =     [5, 1, 0, 10, 20, 500, 850, 700, 300, 100, 20]
Chennai =    [10, 10, 15, 20, 40, 60, 100, 200, 300, 400, 350]
Delhi =      [15, 10, 5, 10, 20, 70, 100, 80, 60, 30, 10]

plt.plot(months, Mumbai, label="Mumbai", color='cyan', linestyle='-', linewidth=2)
plt.plot(months, Chennai, label="Chennai", color='orange', linestyle='--', linewidth=2)
plt.plot(months, Delhi, label="Delhi", color='lime', linestyle='-.', linewidth=2)

plt.xlabel("Month (0 = Jan)")
plt.ylabel("Rainfall (mm)")
plt.title("Average Monthly Rainfall in Indian Cities")
plt.legend()
plt.grid(True)
plt.show()
```
